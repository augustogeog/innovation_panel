## COLLECTING RAIS DATA

In the present notebook, the **download_rais** function from **collect** module was called for 14 designated Brazilian federation unities (FUs), namely AM, BA, CE, DF, ES, GO, MG, PA, PE, PR, RJ, RS, SC and SP. As a result, the compressed files from **[RAIS Microdata Server](ftp://ftp.mtps.gov.br/pdet/microdados/RAIS/)** were donwloaded into rais_original subdirectories. Subsequently, the **extract_rais_original** function was applied to these files, decompressing its .txt extension files and deleting the original compressed ones. For 14 FUs, from 1993 to 2018, the original database comprises 399 files, which occupy 408.5GB of disk memory.

```python
import os
import glob
from app_rais.collect import download_rais
from app_rais.collect import extract_rais_original
```

```python
%%time
download_rais(
    list_ufs=['AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MG','PA','PE','PR','RJ','RS','SC','SP']
    , list_year=list(range(1993,2018))
)
```

```python
%%time
extract_rais_original(
    path_pattern='D:/inovation_panel/inovation_panel/app_rais/data/rais_original/**/**.7z'
    , remove_7z=True
)
```

```python
list_files = glob.glob(f'app_rais/data/rais_original/**/**.txt')
size = round(sum(os.path.getsize(file) for file in list_files) * 1e-9, 2)
print(f'Resulting treated files occupy {size}GB in disk.')
```
