Present notebook calls the **original_to_treated_loop** function from **treat** module for designated Brazilian federation unities (FU - UF in Portuguese). This function: 
1. loads original Rais data from .txt extension files inside rais_original subdirectories;
2. casts dtypes to its columns for better performance;
3. drops columns not relevant for analysis of S&T structure;
4. adds data related to S&T personnel, services according to knowledge intensity and industries technological level;
5. for each FU and year, saves a compressed pickle file in rais_treated subdirectories.

For 14 selected FUs, from 2007 to 2017, saved files sum up to 7GB on disk.

```python
import os
import glob
import app_rais.treat as treat
```

```python
%%time
treat.original_to_treated_loop(
    ufs=['AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MG','PA','PE','PR','RJ','RS','SC','SP']
    , years=list(range(2007,2018))
)
```

```python
list_files = glob.glob(f'app_rais/data/rais_treated/**/**.zip')
size = round(sum(os.path.getsize(file) for file in list_files) * 1e-9, 2)
print(f'Resulting treated files occupy {size}GB in disk.')
```
